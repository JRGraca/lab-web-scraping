#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
````

````python
# your code here
top250_response = requests.get(url)
````

````python
# Get content out of the page
top250_soup = BeautifulSoup(top250_response.content)
items = top250_soup.find_all("td", attrs={"class":"titleColumn"})
ratings = top250_soup.find_all("td", attrs={"class":"ratingColumn imdbRating"})
````

````python
# append to lists and assemble them into a DataFrame

movie_names = []
movie_years = []
movie_directors = []
movie_ratings = []
for i in items:
	movie_names.append(i.find("a").text)
	movie_years.append(i.find("span", attrs={"class":"secondaryInfo"}).text.replace("(","").replace(")",""))
	movie_directors.append(i.find("a")["title"].split(" (dir.)")[0])
for i in ratings:
	movie_ratings.append(i.find("strong").text)


# I couldn't think of a loop for this
top250_table = pd.DataFrame()
top250_table["Title"] = movie_names
top250_table["Release Date"] = movie_years
top250_table["Director"] = movie_directors
top250_table["Rating"] = movie_ratings

top250_table
````

#### 3.1. Display the movie name, year and a brief summary of the top 10 random movies (IMDB) as a pandas dataframe.

````python
#This is the url you will scrape in this exercise
url = 'https://www.imdb.com/list/ls009796553/'
````

````python
# your code here
top10 = requests.get(url)
top10.status_code
````

````python
# Get soup and create empty dataframe
t10_soup = BeautifulSoup(top10.content)
t10_random_df = pd.DataFrame()
````

````python
t10_names = t10_soup.find_all("h3", {"class":"lister-item-header"})
t10_names_l = []
for i in t10_names:
	t10_names_l.append(i.find("a").text)

t10_random_df["Title"] = t10_names_l
````

````python
# returns year
t10_years = t10_soup.find_all("span", attrs={"class":"lister-item-year text-muted unbold"})
t10_years_l = []
for i in t10_years:
	t10_years_l.append(re.sub('[^\d]', '', i.text.replace("(","").replace(")","").replace('[^\d]', "")))
t10_random_df["Year"] = t10_years_l
````

````python
t10_blurbs = t10_soup.find_all("div", attrs={"class":"lister-item-content"})
t10_blurbs_l = []
for i in t10_blurbs:
	t10_blurbs_l.append(i.find("p", attrs={"class":""}).text.strip())
t10_random_df["Description"] = t10_blurbs_l
````
